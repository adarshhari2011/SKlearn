```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.dropna(inplace = True)
```

```python
sns.scatterplot(penguins , x = "body_mass_g" , y = "flipper_length_mm" , hue = "species")
```

```python
X = penguins[["bill_length_mm" , "bill_depth_mm" , "flipper_length_mm"]]
y = penguins["species"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_learn , y_test = train_test_split(X , y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logisticRegressionModel = LogisticRegression()
```

```python
logisticRegressionModel.fit(X_train , y_learn)
```

```python
y_predict = logisticRegressionModel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_predict)
```
